# Mô tả của các côt

`"customer_id"`: Mã định danh của khách hàng

`"label"`: 0: Khách hàng trả nợ đúng hạn
           1: Khách hàng quá hạn trả nợ 

`"_COUNT_"`: Số khoản vay theo từng loại ngắn hạn - trung hạn - dài hạn từ ngân hàng - tổ chức phi ngân hàng 

`"NUMBER_OF_LOANS_"`: Tổng tất cả khoản vay (không phân biệt loại) từ ngân hàng - tổ chức phi ngân hàng

`"NUMBER_OF_CREDIT_CARDS_"`: Tổng số thẻ tín dụng được cấp bởi ngân hàng - tổ chức phi ngân hàng

`"NUMBER_OF_RELATIONSHIP_"`: Tổng số mối quan hệ tài chính mà khách hàng có với ngân hàng - tổ chức phi ngân hàng 

`"NUM_NEW_LOAN_TAKEN_xM"`: Tổng số khoản vay mới từ ngân hàng - tổ chức phi ngân hàng mà khách hàng thực hiện trong 3 - 6 - 9 - 12 tháng trước đó

`"OUTSTANDING_BAL_LOAN_CURRENT"`: Số dư nợ của các khoản vay tính đến hiện tại

`"OUTSTANDING_BAL_LOAN_xM"`: Số dư nợ của các khoản vay trong 3 - 6 - 9 - 12 tháng trước đó

`"OUTSTANDING_BAL_CC_xM"`: Số dư nợ của thẻ tín dụng trong 3 - 6 - 9 - 12 tháng trước đó

`"OUTSTANDING_BAL_ALL_xM"`: Tổng số dư nợ cho tất cả các sản phẩm tài chính trong 3 - 6 - 9 - 12 tháng trước đó

`"OUTSTANDING_BAL_LOAN_xM_yM"`: Số chênh lệch giữa các số dư nợ của các khoản vay trong hai khoảng thời gian

`"OUTSTANDING_BAL_CC_xM_yM"`: Số chênh lệch giữa các số dư nợ của thẻ tín dụng trong hai khoảng thời gian

`"OUTSTANDING_BAL_ALL_xM_yM"`: Số chênh lệch giữa tổng các số dư nợ cho tất cả các sản phẩm tài chính trong hai khoảng thời gian

`"INCREASING_BAL_xM_"`: Số tăng lên trong số dư nợ của các khoản vay - thẻ tín dụng - tất cả sản phẩm tài chính trong 3 - 6 tháng

`"OUTSTANDING_BAL_CC_CURRENT"`: Số dư nợ của thẻ tín dụng tính đến hiện tại

`"CREDIT_CARD_MONTH_SINCE_xDPD"`: Số tháng kể từ khi khách hàng có thanh toán quá hạn 10 -  30 - 60 - 90 ngày trên khoản thanh toán thẻ tín
                                  dụng gần nhất

`"CREDIT_CARD_NUMBER_OF_LATE_PAYMENT"`: Tổng số lần thanh toán trễ của khách hàng trên các thẻ tín dụng

`ENQUIRIES_FROM_FOR_xM`: Số lượt tra cứu tín dụng liên quan đến các khoản vay - thẻ tín dụng từ ngân hàng - tổ chức phi ngân hàng trong 3 - 6 -
                         9 - 12 tháng trước đó

`"ENQUIRIES_FROM_xM_yM"`: Số lượt tra cứu tín dụng liên quan đến các sản phẩm tài chính từ ngân hàng - tổ chức phi ngân hàng trong 2 khoảng thời
                          gian 

`"OUTSTANDING_BAL_ALL_CURRENT"`: Tổng số dư nợ cho tất cả các sản phẩm tài chính tính đến hiện tại

- Ý tưởng: ta fill từ 3M trước giả định rằng nếu trong 3M gần nhất không có giá trị thì khách hàng không yêu cầu thêm nào. Từ 3M ta sẽ fill lên 6M bằng cách nếu trong 6M không có giá trị, ta sẽ cho bằng giá trị từ 3M (giả định rằng từ trong khoảng thời gian từ 3 đến 6 tháng đó khách hàng không yêu cầu thêm), tương tự với 9M và 12M

## Loại bỏ Outliers

```python
def analyze_visualize_outliers(data, num_columns, plot_type="both"):
    """
    Phân tích và trực quan hóa các giá trị ngoại lai trong các cột dữ liệu số của DataFrame.

    Args:
        data (pd.DataFrame): Bộ dữ liệu.
        num_columns (list): Danh sách tên cột để phân tích.

    Returns:
        dict: Danh sách chứa các giá trị ngoại lệ, số lượng và tỷ lệ phần trăm của chúng cho mỗi cột.
    """

    outlier_data = {}
    for col in num_columns:
        data_column = data[col]
        Q1 = data_column.quantile(0.25)
        Q3 = data_column.quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = data_column[(data_column < lower_bound) | (data_column > upper_bound)]
        outlier_counts = len(outliers)
        outlier_percentages = (outlier_counts / len(data_column)) * 100

        outlier_data[col] = {
            'count': outlier_counts,
            'percentage': outlier_percentages,
        }

        print(f"Số lượng IQR outliers trong '{col}': {outlier_counts}")
        print(f"Phần trăm của outliers trong '{col}': {outlier_percentages:.2f}%")
        print("-----------------------------------------------------------------")

    # Vẽ dựa trên biến "plot_type"
    if plot_type == "counts":
        fig, ax = plt.subplots(figsize=(20, 6))
        ax.bar(outlier_data.keys(), [v['count'] for v in outlier_data.values()], color='red')
        ax.set_title('Số lượng Ouliers trong từng cột')
        ax.set_ylabel('Count')
        ax.set_xticks(list(outlier_data.keys())) 
        ax.set_xticklabels(list(outlier_data.keys()), rotation=90)  
            
        plt.tight_layout()
        plt.show()

    elif plot_type == "percentages":
        fig, ax = plt.subplots(figsize=(20, 6))
        ax.bar(outlier_data.keys(), [v['percentage'] for v in outlier_data.values()], color='lightblue')
        ax.set_title('Phần trăm của outliers trong từng cột')
        ax.set_ylabel('Percentage')
        ax.set_xticks(list(outlier_data.keys()))  
        ax.set_xticklabels(list(outlier_data.keys()), rotation=90)  
            
        plt.tight_layout()
        plt.show()

    elif plot_type == "both":
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))

        # Biểu đồ đếm
        ax1.bar(outlier_data.keys(), [v['count'] for v in outlier_data.values()], color='red')
        ax1.set_title('Số lượng Ouliers trong từng cột')
        ax1.set_ylabel('Count')
        ax1.set_xticks(list(outlier_data.keys()))  
        ax1.set_xticklabels(list(outlier_data.keys()), rotation=90)  


        # Biểu đồ %
        ax2.bar(outlier_data.keys(), [v['percentage'] for v in outlier_data.values()], color='orange')
        ax2.set_title('Phần trăm của outliers trong từng cột')
        ax2.set_ylabel('Percentage')
        ax2.set_xticks(list(outlier_data.keys()))  
        ax2.set_xticklabels(list(outlier_data.keys()), rotation=90)


        plt.tight_layout()
        plt.show()

    else:
        print(f"Sai giá trị plot_type: {plot_type}. Các giá trị đúng là 'counts', 'percentages', or 'both'.")

    return outlier_data
```

```python
outlier_df = pd.DataFrame(analyze_visualize_outliers(df.drop(columns='label'),
                                                     df.drop(columns='label').columns, plot_type='percentages'))
```
